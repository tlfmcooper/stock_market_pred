```python
import json
import random

# Define file paths
input_file = './stocks/train_dataset.jsonl'
output_file = './stocks/sample.jsonl'

# Read the original JSONL file
with open(input_file, 'r') as f:
    lines = f.readlines()

# Calculate 10% of the total number of lines
sample_size = int(len(lines) * 0.005)

# Randomly select 10% of the data
sampled_lines = random.sample(lines, sample_size)

# Write the sampled data to a new JSONL file
with open(output_file, 'w') as f:
    for line in sampled_lines:
        f.write(line)
```
