```python
import json

# Define the path to your JSONL file
jsonl_file_path = ".\stocks\instruction_dataset.jsonl"

# Read and display the first 10 entries from the JSONL file
with open(jsonl_file_path, 'r') as f:
    for i, line in enumerate(f):
        if i >= 10:  # Limit to the first 10 entries
            break
        entry = json.loads(line)
        print(f"Entry {i + 1}: {entry}\n")
```
